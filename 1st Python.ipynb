```python
#!/usr/bin/env python
# coding: utf-8

# # Python Fundamentals

# ### Print in Python

# In[11]:


print('Hello World')


# In[12]:


## Single Line Comment - comment in python
## Multi -Line Comment

"""
This is a comment
written in
more than just one line
"""
print('hello')


# In[16]:


# Program to print output in a program
salary = 7000 ##variable in python is called an identifier
print("Total Salary is salary")
print("Total Salary is : ",salary)


# In[23]:


#Program to use "end" argument | "end" helps to print multiple argument in a single line and 
#cause curson to remain in the same line
print('Welcome',end=' ')
print('to',end=' ')
print('Python',end=' ')


# In[30]:


#Program to show the use of "sep" argument
# The print function use a keyword "sep" to specify the string use to insert b/w items.
var1, var2, var3 = 100, 200, 300
#without sep
print(var1, var2, var3)
#with sep ","
print(var1, var2, var3, sep=',')


# ### Input in Python

# In[31]:


# By Default input function helps to take the string data from the users.
# However int() and float() can be used along the input function if user want to take only the int and float.
# eval() is commonly used function that defines the type of input according to the value entered.
# type() function is used to identify the datatype.


# In[33]:


# input function
num = input("Enter the num : ") #by default input takes string
print("the number is ",num," and the type is ",type(num))


# In[34]:


# input function with int
num = int(input("Enter the number:")) #int give a whole number
print("The number is ",num,"and the type is ",type(num))


# In[35]:


# input function with float
num = float(input("Enter the number:"))#float give a decimal no with precison 1
print("The number is ",num,"and the type is ",type(num))


# In[39]:


# input with eval
# eval function considers the datatype according to the input provided by the user.
var1 = eval(input("enter the value of the first variable :"))
print('Var1=',var1,"Type =", type(var1))

var2 = eval(input("enter the value of the Second variable :"))
print('Var2=',var2,"Type =", type(var2))

var3 = eval(input("enter the value of the third variable :"))
print('Var3=',var3,"Type =", type(var3))


# ### Accepting multiple inputs by the users

# In[41]:


# Program to ask multiple inputs by the users
a,b,c = eval(input("Enter the three number:"))
print(a,"*",b,"*",c,"=",a*b*c)


# In[ ]:




```
