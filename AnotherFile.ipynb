```python
# Asking user for Input
Choice = input("Enter your choice of Pangram, Consogram and Vowgram :")

 


# Defining the letters for checking in different scenarios
if Choice == "Pangram":# Logical operator to check the condition
    letters = "abcdefghijklmnopqrstuvwxyz" # assignment operator to assign if the value is true in logical operator
elif Choice == "Consogram":
    letters = "bcdfghjklmnpqrstwxyz"
elif Choice == "Vowgram":
    letters = "aeiou"
else:
    print("Not a Valid Choice")
    

 

# Taking the input from user
User_input = input("Enter the string :")
flag = True

 

# Checking each and every letter in the string
for letter in letters:
    if letter not in User_input.lower():
        flag=False
        break
        
# Displaying the output
if flag == False:
    print("Not a ", Choice)
else:
    print("It is a ", Choice)
```
